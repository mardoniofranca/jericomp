```python
import numpy as np
import matplotlib.pyplot as plt

# Função para simular o movimento de um caminhante aleatório
def caminhante_aleatorio(steps):
    # Inicializa a posição no zero
    position = 0
    positions = [position]

    # Para cada passo, atualiza a posição com +1 ou -1 aleatoriamente
    for step in range(steps):
        move = np.random.choice([-1, 1])
        position += move
        positions.append(position)

    return positions

# Número de passos
n_steps = 1000

# Simula o caminhante aleatório
positions = caminhante_aleatorio(n_steps)

# Gráfico da trajetória do caminhante
plt.figure(figsize=(10, 6))
plt.plot(positions, color='blue', alpha=0.7)
plt.title('Trajetória do Caminhante Aleatório (1D)')
plt.xlabel('Número de Passos')
plt.ylabel('Posição')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Função para simular o movimento de um caminhante aleatório
def caminhante_aleatorio(steps):
    # Inicializa a posição no zero
    position = 0

    # Para cada passo, atualiza a posição com +1 ou -1 aleatoriamente
    for step in range(steps):
        move = np.random.choice([-1, 1])
        position += move

    return position

# Parâmetros
n_steps = 1000  # Número de passos por simulação
n_simulations = 5000  # Número de vezes que o caminhante será simulado

# Executa múltiplas simulações e registra as posições finais
final_positions = [caminhante_aleatorio(n_steps) for _ in range(n_simulations)]

# Calcula a distribuição de probabilidades
bins = np.arange(-n_steps, n_steps + 1, 2)  # Definir bins para o histograma
prob, edges = np.histogram(final_positions, bins=bins, density=True)

# Gráfico da distribuição de probabilidades
plt.figure(figsize=(10, 6))
plt.bar(edges[:-1], prob, width=2, color='blue', alpha=0.7)
plt.title(f'Distribuição Gaussiana - Caminhante Aleatório ({n_simulations} Simulações)')
plt.xlabel('Posição Final')
plt.ylabel('Probabilidade')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Função para simular o movimento de um caminhante aleatório
def caminhante_aleatorio(steps, n_walkers):
    # Inicializa as posições de todos os caminhantes no zero
    positions = np.zeros(n_walkers)

    # Para cada passo, atualiza as posições com +1 ou -1 aleatoriamente
    for step in range(steps):
        steps = np.random.choice([-1, 1], size=n_walkers)
        positions += steps

    return positions

# Número de passos e número de caminhantes
n_steps = 1000
n_walkers = 5000

# Simula o caminhante aleatório
final_positions = caminhante_aleatorio(n_steps, n_walkers)

# Calcula a distribuição de probabilidades
bins = np.arange(-n_steps, n_steps+1, 2) # Define os bins para o histograma
prob, edges = np.histogram(final_positions, bins=bins, density=True)

# Gráfico da distribuição de probabilidades
plt.figure(figsize=(10, 6))
plt.bar(edges[:-1], prob, width=2, color='blue', alpha=0.7)
plt.title('Distribuição de Probabilidades do Caminhante Aleatório')
plt.xlabel('Posição Final')
plt.ylabel('Probabilidade')
plt.grid(True)
plt.show()
```
