```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
# 1. Gerar os dados
np.random.seed(0)
X = np.linspace(0, 2 * np.pi, 100)
y = np.sin(X)

# Redimensionar para o formato (n amostras, 1) que o sklearn espera
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

# 2. Criar o modelo de regressão polinomial (grau 5, por exemplo)
grau = 5
modelo = make_pipeline(PolynomialFeatures(grau), LinearRegression())
```

```python
# 3. Treinar o modelo
modelo.fit(X, y)
```

```python
# 4. Fazer previsões
X_test = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)
y_pred = modelo.predict(X_test)
```

```python
# 5. Plotar os resultados
plt.figure(figsize=(10, 6))
plt.plot(X, y, label="Função $\sin(x)$", color='blue')
plt.plot(X_test, y_pred, label=f"Regressão polinomial (grau {grau})", color='red', linestyle='--')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Aproximação de $\sin(x)$ com Regressão Polinomial")
plt.grid(True)
plt.show()
```

```python
# 1. Gerar os dados
np.random.seed(0)
X = np.linspace(0.1, 10, 100)  # evita x = 0 para não dividir por zero
y = np.sin(X) / X

X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
# 2. Criar modelo de regressão polinomial
grau = 7  # você pode testar outros valores
modelo = make_pipeline(PolynomialFeatures(grau), LinearRegression())

# 3. Treinar o modelo
modelo.fit(X, y)
```

```python
# 4. Fazer previsões
X_test = np.linspace(0.1, 10, 200).reshape(-1, 1)
y_pred = modelo.predict(X_test)

# 5. Plotar os resultados
plt.figure(figsize=(10, 6))
plt.plot(X, y, label=r"$\frac{\sin(x)}{x}$", color='blue')
plt.plot(X_test, y_pred, label=f"Regressão polinomial (grau {grau})", color='red', linestyle='--')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Aproximação de $\frac{\sin(x)}{x}$ com Regressão Polinomial")
plt.grid(True)
plt.show()
```
