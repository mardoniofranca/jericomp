```python
import numpy as np
import matplotlib.pyplot as plt
from particula import Particula

plt.rcParams['font.size'] = 12
```

```python
# Instancia Part√≠cula
p1 = Particula(0,   0,    # (x0, y0) = (0, 0)
              10*np.cos(np.deg2rad(15)), 10*np.sin(np.deg2rad(15)),   # (vx, vy) = (0, 0)
              1)         # m = 1kg

p2 = Particula(0,   0,    # (x0, y0) = (0, 0)
               10*np.cos(np.deg2rad(30)), 10*np.sin(np.deg2rad(30)),  # (vx, vy) = (0, 0)
               1)         # m = 1kg

p3 = Particula(0,   0,    # (x0, y0) = (0, 0)
               10*np.cos(np.deg2rad(45)), 10*np.sin(np.deg2rad(45)),  # (vx, vy) = (0, 0)
               1)         # m = 1kg

p4 = Particula(0,   0,    # (x0, y0) = (0, 0)
               10*np.cos(np.deg2rad(60)), 10*np.sin(np.deg2rad(60)),  # (vx, vy) = (0, 0)
               1)         # m = 1kg


# parametrizacao
dt = 0.01     # intervalo de tempo
tempo = 200   # tempo total

# evaluate data
for i in range(int(tempo/dt)):
  for p in [p1, p2, p3, p4]:
    p.newton(0, -9.8, dt)
```

```python
# plot
for theta, p in zip(["15", "30", "45", "60"],
                    [ p1 ,  p2 ,  p3, p4]):
  plt.plot(p.x, p.y, lw=2, label=rf'$\theta = {theta}^\circ$')

plt.legend()
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.grid(lw=0.5)
plt.show()
```
